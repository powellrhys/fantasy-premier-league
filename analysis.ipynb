```python
df = pd.read_csv('data_historic/historic_player_data.csv').drop('index', axis=1)
```

```python
columns = ['season_x', 'name', 'position', 'assists', 'bonus', 'bps', 'clean_sheets', 
           'goals_conceded', 'goals_scored', 'minutes', 'own_goals', 'penalties_missed', 
           'penalties_saved', 'red_cards', 'saves', 'yellow_cards', 'value', 'total_points']

df = df[columns]
df.head()
```

```python
season = df.groupby(['season_x']).sum()
season
```

```python
position = df.groupby(['position']).sum()
position
```

```python
player = df.groupby(['season_x', 'name', 'position']).sum().sort_values(by='total_points', ascending=False)
player.head(40)
```
